```python
import pandas as pd

df = pd.read_csv("data.csv")
df.head()
```

```python
df.info()
df.describe()
```

Data Cleaning

```python
df.isnull().sum()
```

for filling the missing values in numeric column

```python
df['sales'] = df['sales'].fillna(df['sales'].mean())
```

Data Transformation

```python
df_encoded = pd.get_dummies(df, columns=['city','category'])

df_encoded.head()
```

Data Visualization

```python
import matplotlib.pyplot as plt

df['sales'].hist()
plt.title("Sales Distribution")
plt.xlabel("Sales")
plt.ylabel("Frequency")
plt.show()
```

Data Visualization Part 2 :-

```python
plt.scatter(df['age'], df['sales'])
plt.xlabel("Age")
plt.ylabel("Sales")
plt.title("Age vs Sales")
plt.show()
```
